```python
#Using item similarity 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
columns_names = ['user_id', 'item_id', 'rating', 'timestamp']
```

```python
df = pd.read_csv('u.data', sep='\t', names=columns_names)
```

```python
df.head()
```

```python
movie_titles = pd.read_csv('Movie_Id_Titles')
```

```python
movie_titles.head()
```

```python
df = pd.merge(df, movie_titles, on='item_id')
```

```python
df.head()
```

```python
sns.set_style('white')
```

```python
#Ratings dataframe with average rating and number of ratings
df.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
df.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
```

```python
ratings.head()
```

```python
ratings['number of ratings'] = df.groupby('title')['rating'].count()
```

```python
ratings.head()
```

```python
plt.figure(figsize=(10,5))
ratings['number of ratings'].hist(bins=70)
```

```python
plt.figure(figsize=(10,5))
ratings['rating'].hist(bins=70)
```

```python
plt.figure(figsize=(10,5))
sns.jointplot(x='rating', y='number of ratings', data=ratings, alpha=0.5)
```

```python
moviemat = df.pivot_table(index='user_id', columns='title', values='rating')
```

```python
moviemat.head()
```

```python
ratings.sort_values('number of ratings', ascending=False).head(10)
```

```python
starwars_user_ratings = moviemat['Star Wars (1977)']
liarliar_user_ratings = moviemat['Liar Liar (1997)']
```

```python
starwars_user_ratings.head()
```

```python
liarliar_user_ratings.head()
```

```python
similar_to_starwars = moviemat.corrwith(starwars_user_ratings)
```

```python
similar_to_liarliar = moviemat.corrwith(liarliar_user_ratings)
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars, columns = ['Correlation'])
corr_starwars.dropna(inplace=True)
```

```python
corr_starwars.head()
```

```python
corr_starwars.sort_values('Correlation', ascending=False).head()
```

```python
corr_starwars = corr_starwars.join(ratings['number of ratings'])
```

```python
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['number of ratings']>100].sort_values('Correlation',
                                                                 ascending=False).head()
```

```python
#Trying the same on Liar Liar
corr_liarliar = pd.DataFrame(similar_to_liarliar, columns=['Correlation'])
```

```python
corr_liarliar.dropna(inplace=True)
```

```python
corr_liarliar = corr_liarliar.join(ratings['number of ratings'])
```

```python
corr_liarliar
```

```python
corr_liarliar[corr_liarliar['number of ratings']>100].sort_values('Correlation',
                                                                 ascending=False).head()
```
