```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
print(boston.DESCR)
boston_df = boston.data
```

```python
print(boston['target'])
```

```python
boston['feature_names']
```

```python
df = pd.DataFrame(boston.data, columns=boston['feature_names'])
```

```python
df['target'] = boston.target
```

```python
#Splitting data into X array (contains features to train on) & y array (contains target variable)

X = df[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']]
y = df['target']
```

```python
#Train Test Split
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
#Creating and training the model
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

```python
#Model Evaluation
print(lm.intercept_)
```

```python
cdf = pd.DataFrame(lm.coef_, X.columns, columns=['Coefficients'])
```

```python
cdf
```

```python
# With 1 unit of increase in per capita crime rate by town (CRIM), the pricing of houses in the area decreases by 0.077$.

# With 1 unit of increase in proportion of residential land zoned for lots over 25,000 sq.ft (ZN), the pricing of houses
# increases by 0.04$.

# With 1 unit of increase in proportion of non-retail business acres per town (INDUS), the pricing of houses increases by 0.09$.

# With 1 unit of increase in nitric oxides concentration (parts per 10 million) (NOX), the pricing of houses decreases by 19.9$.

# With 1 unit of increase in average number of rooms per dwelling (RM), the pricing of houses increases by 2.89$.
```

```python
#Predictions from Model

predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test, predictions)
```

```python
# Residual Histogram
sns.distplot((y_test - predictions), bins=50) 
```

```python
#Regression Evaluation Metrics
from sklearn import metrics
```

```python
print("MAE: ", metrics.mean_absolute_error(y_test, predictions))
```

```python
print("MSE: ", metrics.mean_squared_error(y_test, predictions))
```

```python
print("RMSE: ", np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
